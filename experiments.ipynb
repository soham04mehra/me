```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
import pickle
```

```python
#file got read by pandas
my_file=pd.read_csv('Churn_Modelling.csv')
my_file.head()
```

```python
#dropping values
data_drop=my_file.drop(['RowNumber','CustomerId','Surname'],axis=1)
data_drop.head()
```

```python
label_encoder=LabelEncoder()
data_drop['Gender']=label_encoder.fit_transform(data_drop['Gender'])
data_drop.head()
```

```python
#one hot encoded the geography and got the features out 
hot_encoder=OneHotEncoder()
one_hot_encoded=hot_encoder.fit_transform(data_drop[['Geography']]).toarray()
sep_names=hot_encoder.get_feature_names_out(['Geography'])
one_hot_encoded
sep_names
```

```python
#making a data frame of both and merging this with old one
new_data=pd.DataFrame(one_hot_encoded,columns=sep_names)
new_data
```

```python
result=pd.concat([data_drop.drop(['Geography'],axis=1),new_data],axis=1)
result.head()
```

```python
import pickle
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder,file)
with  open('hot_encoder.pkl','wb') as file:
    pickle.dump(hot_encoder,file)
```

```python
#splitiing into independent and dependent features
x=result.drop(['Exited'],axis=1)
y=result['Exited']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
standard_scalar=StandardScaler()
x_train=standard_scalar.fit_transform(x_train)
x_test=standard_scalar.transform(x_test)
```

```python
import pickle
with open('standard_scaler.pkl','wb') as file:
    pickle.dump(standard_scalar,file)
```

IMPLEMENTATION  ON THE MODEL

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1]
```

```python
#creating hidden layes
model=Sequential([
Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
Dense(32,activation='relu'),
Dense(1,activation='sigmoid')
])
```

```python
#compling the model
import tensorflow as tf
loss=tf.keras.losses.BinaryCrossentropy()
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#setting up tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#set up early stop
Early_Stopping=EarlyStopping(monitor='val-loss',patience=10,restore_best_weights=True)

```

```python
#train the model 
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,Early_Stopping]
)
```

```python
model.save('model.h5')
```
